```python
from sklearn.cluster import DBSCAN

from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt

from sklearn.datasets import make_moons

from sklearn.preprocessing import StandardScaler

from sklearn.preprocessing import MinMaxScaler

import numpy as np

import pandas as pd

from scipy.spatial.distance import pdist, squareform
```

```python
#input_spot = int(input())
#input_range = float(input())
input_spot = 500
input_range = 10


np.random.seed(111)
#np.random
var = ['X', 'Y']
labels=['spot0']
for i in range(0,input_spot-1):
    z = 'spot'+str(i+1)
    labels.append(z)

#print(labels)
X = np.random.random_sample([input_spot,2])*(input_spot*2)
df = pd.DataFrame(X, columns = var, index = labels)
#df는 점5개의 x,y 좌표를 저장해둠

distmatrix = pdist(df, metric = 'euclidean')
row_dist = pd.DataFrame(squareform(distmatrix), columns = labels, index = labels)

test = df.values
```

```python
scaler_ss = StandardScaler().fit(test)

x_scaled_ss = scaler_ss.transform(test) 

dbscan = DBSCAN(eps=0.18, min_samples=6)

clusters_ss = dbscan.fit_predict(x_scaled_ss)

print('np.unique(clusters_ss) 예측한 레이블:{}'.format(np.unique(clusters_ss))) # [0 ,1]
### 2차원 데이터셋을 0과 1로 구분했기 때문에 전처리가 잘되었음을 확인

length = len(format(np.unique(clusters_ss)))



# visualization

df = np.hstack([test, clusters_ss.reshape(-1, 1)]) # x_scaled_ss 오른쪽에 1열 붙이기

for i in range(length):
    df_ft = df[df[:,2]==i, :]
    plt.scatter(df_ft[:, 0], df_ft[:, 1], cmap='tab20b')

# 노이즈 출력
df_ft1 = df[df[:,2]==-1, :]
plt.scatter(df_ft1[:, 0], df_ft1[:, 1], color = 'black')

ar = []
for i in range(len(df)):
    if(df[i,2]==-1):
        ar.append(i)
print(ar)


plt.grid(True)
plt.show()
```
